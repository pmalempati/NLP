```python
url="https://www.cnbc.com/2019/01/17/netflix-price-hike-helps-disney-upcoming-streaming-service-analyst.html"
```

```python
html=request.urlopen(url).read()
```

```python
raw=BeautifulSoup(html)
```

```python
for script in raw(["script","style"]):
    script.extract()
    
text= raw.get_text()

#lines=(line.strip() for line in text.splitlines())
#chunks=(phrase.strip()for line in lines for phrase in line.split(" "))
#text='\n'.join(chunk for chunk in chunks if chunk)
print(text)
```

```python
matches=re.findall(r'\$\w+',text)
```

```python
print(matches)
```

```python
result=re.sub('\w*[0-9]\w*','#',text)
```

```python
print(result)
```

```python
countNet=text.count("Netflix")
print(countNet)
countDis=text.count("Disney")
print(countDis)
```
